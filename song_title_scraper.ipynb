## Song titles scrapping

```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
from time import sleep
import csv
import re
```

```python
# url to scrape the songs list from
base_url = "https://www.azlyrics.com/a/{}.html"
artists = ["arianagrande","annmarie","aaronlewis","acidghost","adultmom","adventureclub","after7","asia","activechild","avatar",
           "atlasgenius","autoheart","ananyabirla","architects","aprilwine","aaradhna","aarondoh","aaronneville",
           "aaronwatson","abc","accept"]
songs_dict = { }    
```

```python
for artist in artists:
    
    artist_url = base_url.format(artist)
    print("Going to url : ", artist_url)
    
    headers = {'User-Agent': 'Mozilla/5.0'}
    req = Request(artist_url, headers=headers) 
    html_page = urlopen(req) 

    #html_page = urlopen(artist_url)
    soup = BeautifulSoup(html_page, 'html.parser')
    
    songs_list = soup.find('div', attrs={'id':'listAlbum'}).find_all('div', attrs={'class':'listalbum-item'})
    songs_dict[artist] = []
    
    for song in songs_list:
        song_name = song.text.strip()
        numbers_in_brackets_removed = re.sub(r'\(.*\)',"",song_name)
        processed_song = re.sub(r'\W+', '', numbers_in_brackets_removed).lower()
        songs_dict[artist].append(processed_song)
        
    print("Artist : ", artist)
    print(songs_dict[artist][:5])
        
    sleep(10)
```

```python
for key,val in songs_dict.items():
    for i in range(0,len(val)):
        print(key,val[i])
```

```python
import json
json_file = "new_artist_songs_list.json"
with open(json_file, 'w') as file:
    json.dump(songs_dict, file)
```
